This script is to find out what species don't have neuron-specific genes as the lowest pairwise dN/dS genes against human reference genome. 

```python
import numpy as np
import pandas as pd
```

```python
dnds_df=pd.read_excel('../results/celltype-specific_descriptive_stats.xlsx',sheet_name='human reference genome')
```

```python
print("Species with lowest dN/dS cell type other than neuron:")
for spp in dnds_df['SpeciesCommonName'].drop_duplicates():
    df=dnds_df[dnds_df['SpeciesCommonName']==spp]
    if df[df['Celltype']=='neuron']['Median_dNdS'].values > df['Median_dNdS'].T.values.min():
        print(spp)
```

```python
print("Species with neuron-specific genes' dNdS significantly higher than some other cell type:")
for spp in dnds_df['SpeciesCommonName'].drop_duplicates():
    df=dnds_df[dnds_df['SpeciesCommonName']==spp]
    if df[df['Celltype']=='neuron']['Median_dNdS'].values > df['Confidence_Interval_High'].T.values.min():
        print(spp)
```

```python
print("Species with neuron-specific genes median dNdS greater or within confidence interval of some other species:")
for spp in dnds_df['SpeciesCommonName'].drop_duplicates():
    df=dnds_df[dnds_df['SpeciesCommonName']==spp]
    if df[df['Celltype']=='neuron']['Median_dNdS'].values > df[df['Celltype']!='neuron']['Confidence_Interval_Low'].T.values.min():
        print(df)
```
